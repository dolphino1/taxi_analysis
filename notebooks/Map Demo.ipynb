```python
import matplotlib.pyplot as plt
import geopandas as gpd

FILE_PATH = "geojson/manhattan_zip_codes.geojson"

df = gpd.read_file(FILE_PATH)

print(type(df))
```

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (100, 60)


df.plot()
plt.show()
```

```python
import json
with open(FILE_PATH) as data_file:    
    data = json.load(data_file)
```

```python
data['features'][1]['properties']['borough']
new_features = []

for shape in data['features']:
    if shape['properties']['borough'] == 'Manhattan':
        new_features.append(shape)
new_data = data
new_data['features'] = new_features 
```

```python
new_data
```

```python
with open('manhattan_zip_codes.geojson', 'w') as outfile:
    json.dump(new_data, outfile)
```
